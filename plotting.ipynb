# Plotting

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(
    {'vocab_size': ['21K', '30K', '40K', '50K', '60K'],
     'BLEU_test': [19.58 , 19.207, 19.726, 19.556, 19.404]
    }
)
plt.figure(figsize=(10, 6))
ax = df.plot.bar(x='vocab_size', rot=0)
ax.set_xlabel('Size of Japanese Vocabulary', size=13)
ax.set_ylabel('BLEU', size=15)
ax.legend([])
plt.title('Comparing the BLEU scores for machine translation models (Japanese to English) \n trained with different sizes of Japanese vocabulary', y=1.05)
plt.savefig('bleu_scores.png', bbox_inches='tight')
```
